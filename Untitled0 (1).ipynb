```python
import requests
from bs4 import BeautifulSoup

def extract_text_from_url(url):
    response = requests.get(url)
    response.encoding = 'utf-8'
    soup = BeautifulSoup(response.text, "html.parser")

    # Supprimer les balises inutiles
    for tag in soup(["script", "style", "header", "footer", "nav", "form"]):
        tag.decompose()

    # Extraire le texte brut
    text = soup.get_text(separator='\n')
    lines = [line.strip() for line in text.splitlines() if line.strip()]
    return "\n".join(lines)

# Dictionnaire des URLs
urls = {
    "masters": "https://www.fsdm.usmba.ac.ma/Formations/Master",
    "candidature": "https://www.fsdm.usmba.ac.ma/Acces/Candidature2025"
}

# Extraction et affichage du texte brut
for name, url in urls.items():
    print(f"\n📄 Contenu extrait de {name.upper()}:\n{'='*40}")
    raw_text = extract_text_from_url(url)
    print(raw_text)  # Affiche les 3000 premiers caractères (tu peux modifier)
    print(f"\n🔚 Fin de l'extrait de {name.upper()}\n{'='*40}\n")
```
